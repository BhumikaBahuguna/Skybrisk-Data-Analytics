# Customer Churn EDA

This notebook performs exploratory data analysis on the Telco Customer Churn dataset, including cleaning, feature engineering, encoding, and business-focused visualizations.

```python
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

PROJECT_ROOT = Path.cwd().resolve().parent
SCRIPT_DIR = PROJECT_ROOT / "scripts"
if str(SCRIPT_DIR) not in sys.path:
    sys.path.append(str(SCRIPT_DIR))

from churn_cleaning import prepare_churn_data

sns.set_theme(style="whitegrid")
```

```python
raw_data_path = PROJECT_ROOT / "WA_Fn-UseC_-Telco-Customer-Churn.csv"
eda_df, encoded_df = prepare_churn_data(raw_data_path)

print("EDA dataset shape:", eda_df.shape)
print("Encoded dataset shape:", encoded_df.shape)
eda_df.head()
```

```python
eda_df.info()
```

```python
eda_df[["tenure", "MonthlyCharges", "TotalCharges", "avg_monthly_spend"]].describe()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=eda_df, x="Churn", palette="Set2")
plt.title("Customer Churn Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=eda_df, x="Contract", hue="Churn", palette="Paired")
plt.title("Contract Type vs Churn")
plt.xticks(rotation=15)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
correlation_df = encoded_df.select_dtypes(include="number")
top_corr = correlation_df[["Churn", "tenure", "MonthlyCharges", "TotalCharges", "avg_monthly_spend", "SeniorCitizen"]].corr()
sns.heatmap(top_corr, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=eda_df, x="Churn", y="MonthlyCharges", palette="Set3")
plt.title("Monthly Charges by Churn")
plt.show()
```
